**Bar Plot**

```python
import matplotlib.pyplot as plt;

x = ["python", "c", "c++", "java"];
y = [85, 70, 60, 82]; # Popularities
plt.xlabel("Languages", fontsize = 20); # X-axis ke along label lagane ke liye.
plt.ylabel("Popularity", fontsize = 20); # Y-axis ke along label lagane ke liye.
plt.title("Nodix's Workspace", fontsize = 30); # Apne Graph ko title dene ke liye.
color = ["y", "g", "b", "m"]; # Yeh humne color ke liye ek lists bna di ab isse bar plot mei bhejenge.
plt.bar(x,y, width = 0.4, color=color, edgecolor = "black", linewidth = 3, alpha = 0.5);
plt.show();
```
